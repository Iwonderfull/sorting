````python
class AnimalGenerator:
    def __init__(self, seq_end: int):
        self._end = int(seq_end)
        self._state = -1
    
    def __iter__(self):
        return self

    def __next__(self):
        self._state += 1
        value = self.multiply()
        if self.is_empty:
            raise StopIteration()
        return (value, value % 2 == 0)

    def multiply(self):
        return self._state * 2
    
    @property
    def is_empty(self):
        return self._state >= self._end - 1

    def reset(self):
        self._state = -1
````
